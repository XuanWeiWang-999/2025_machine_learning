## Algorithm for Q1.

**Input:** parameters $\{W^{l}, b^{l}\}_{l=2}^{L}$, input $x$.  
**Output:** $\nabla_x a^{[L]}(x)$

#### 1. Forward Pass
1. $a^{[1]}\leftarrow x$
2. for $l=2,\dots,L$
    - $z^{[l]}\leftarrow W^{[l]}a^{[l-1]}+b^{[l]}$
    - $a^{[l]}\leftarrow \sigma\big(Z^{[l]}\big)$

#### 2. Backward pass (output gradient)
1. $g^{[L]}\leftarrow\sigma'\big(z^{[L]}\big)$
2. for $l'=L-1,\dots,2$:
    - $g^{[l]}\leftarrow D^{[l]}\big(W^{[l+1]}\big)^{\top}g^{[l+1]}$

#### Recall the gradient
$\nabla_{x}a^{[l]}\leftarrow \big(W^{[2]}\big)^{\top}g^{[2]}$

---

**Compact closed form:**
$$
\nabla_x a^{[L]}(x)
= \big(W^{[2]}\big)^{\!\top} D^{[2]} \big(W^{[3]}\big)^{\!\top} D^{[3]}
\cdots \big(W^{[L]}\big)^{\!\top} D^{[L]} \cdot \mathbf{1},
$$
where $n_L=1$ so $\mathbf{1}=1$; if the last layer is linear ($\sigma=\mathrm{id}$), then $D^{[L]}=1$.
