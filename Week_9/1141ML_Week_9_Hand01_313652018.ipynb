# Handwritting assignment 01


*Pf.*  
  
Consider a 1-D SDE 
$$
dx_t = f(x_t, t)\,dt + g(x_t, t)dW_t,
$$ 
the pdf. $p(x,t)$ satisfies Fokker-Plank equation 
$$
\partial_t p=-\partial_x(fp)+\frac{1}{2}\partial_{xx}(g^2p).
$$


The RHS can be sort up as $\partial_t p=-\partial_x(u\,p)$. We rewite the second term of the RHS and get 
$$
\frac{1}{2}\partial_{xx}(g^2 p)=\frac{1}{2}\partial_x((\partial_x g^2)p+g^2\partial_x p)=\frac{1}{2}\partial_x((\partial_x g^2)p+g^2p\partial_x\log p)
$$

Hence, 
$$
\partial_t p=-\partial_x(fp)+\frac{1}{2}\partial_x((\partial_x g^2)p+g^2p\partial_x\log p)=-\partial_x[(f-\frac{1}{2} \partial_x g^2-\frac{g^2}{2}\partial_x\log )p]
$$

Therefore, we have 
$$
u(x,t)=f(x,t)-\frac{1}{2}\partial_x g^2(x,t)-\frac{g^2(x,t)}{2}\partial_x\log p(x,t).
$$

Finally, the corresponding probability flow ODE is 
$$
dx_t = \left[f(x_t,t) -\frac{1}{2}\frac{\partial}{\partial x}g^2(x_t, t)- \frac{g^2(x_t,t)}{2}\frac{\partial}{\partial x}\log p(x_t,t)\right]\,dt.
$$

$\square$
